# Minimum working example for visualizing street and bike networks

```python
import matplotlib.pyplot as plt
import osmnx as ox
print(ox.__version__)
print('highway' in ox.settings.useful_tags_way)
cf = ['["highway"~"cycleway"]']


fig = plt.figure(figsize=(4,4))
ax = fig.add_axes([0,0,1,1])


# Streets
g = ox.graph_from_place(
    "Budapest", network_type="all", retain_all=True
)
g_undir = g.to_undirected().copy()
_, edges = ox.graph_to_gdfs(g_undir)
edges = edges.to_crs('3857')
edges.plot(ax=ax, color="k");


# highway~cycleway
g_bike = ox.graph_from_place(
    "Budapest", custom_filter=cf, retain_all=True
)
g_bike_undir = g_bike.to_undirected().copy()
_, edges_bike = ox.graph_to_gdfs(g_bike_undir)
edges_bike = edges_bike.to_crs('3857')
edges_bike.plot(ax=ax, color="r");
```
